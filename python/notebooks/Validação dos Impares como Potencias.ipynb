```python
import numpy as np
import pandas as pd
```

```python
def geraArray(numero, impar = True):
    
    end = False
    curr = 1 if impar else 0
    array = list()
    
    while end != True:
        if len(array) != numero:
            array.append(curr)
            curr += 2
        else:
            end = True
            
    return np.array(array)    
```

```python
a = pd.DataFrame({i:(geraArray(i).sum(),i**2) for i in np.arange(1,20000)}).T
```

```python
a.rename(columns={0:'impares',1:'Real'})
```
